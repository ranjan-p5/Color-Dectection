```python
import cv2
import pandas as pd
img=r'C:\Users\ADMIN\Downloads\color_spectrum.jpg'
csv_file=r'C:\Users\ADMIN\Downloads\colors.csv'
index=['color','color_names','hex','r','g','b']
data=pd.read_csv(csv_file,names=index,header=None)
img_read=cv2.imread(img)
if img_read is None:
    print("Error: Could not read the image. Check your file path!")
else:
    print("Image loaded successfully!")
    img_read=cv2.resize(img_read,(500,500))
clicked=False
r=g=b=xposition=yposition=0
def color_name(r,g,b):
    mini=10000
    for n in range(len(data)):
        d=abs(r-int(data.loc[n,'r']))+abs(g-int(data.loc[n,'g']))+abs(b-int(data.loc[n,'b']))
        if d<=mini:
            mini=d
            cname=data.loc[n,'color_names']
    return cname
def draw_function(event,x,y,flag,params):
    if event==cv2.EVENT_LBUTTONDBLCLK:
        global r,b,g,xposition,yposition,clicked
        clicked=True
        xposition=x
        yposition=y
        b,g,r=img_read[y,x]
      
        b=int(b)
        g=int(g)
        r=int(r)
        
cv2.namedWindow('image')
cv2.setWindowProperty('image', cv2.WND_PROP_TOPMOST, 1)
cv2.setMouseCallback('image',draw_function)
while True:
    cv2.imshow('image',img_read)
    if clicked:
        cv2.rectangle(img_read, (20,20), (600,60), (b,g,r), -1)
        text=color_name(r,g,b)+'R='+str(r)+'G='+str(g)+'B='+str(b)
        cv2.putText(img_read,text,(60,60),2,0.8,(255,255,255),2,cv2.LINE_AA)
        if r+g+b >=600:
            cv2.putText(img_read,text,(60,60),2,0.8,(0,0,0),2,cv2.LINE_AA)
        clicked=False    
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()       
            
```
